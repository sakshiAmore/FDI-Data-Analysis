### Sector-wise Analysis:

```python
#1.Which sectors received the highest FDI inflows on average over the 17-year period?
pd1=pd
# Create a new column 'Sum' in pd2 which is the sum of sales data across different years
pd1['Sum'] = pd[['2000-01', '2001-02', '2002-03', '2003-04', '2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11', '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']].sum(axis=1)# Calculate the average of the sum values across rows
pd1['Avg'] = pd1['Sum'] / 17
```

```python
pd1[pd1['Avg']==pd1['Avg'].max()]
#'SERVICES SECTOR' received the highest FDI inflows on average over the 17-year period"
```

```python
#How did FDI inflows vary across different sectors each year? Are there any noticeable trends?
#here the difference is showen on the bases of color
#this is analysis of 2000-01
plt.figure(figsize = (5, 100)) 
sns.countplot(y='Sector',hue = '2000-01',data=pd)
```

```python
#this is analysis of 2001-02
plt.figure(figsize = (5, 100)) 
sns.countplot(y='Sector',hue = '2001-02',data=pd)
```

```python
#this is analysis of 2016-17 after 17 years
plt.figure(figsize = (5, 100)) 
sns.countplot(y='Sector',hue = '2016-17',data=pd)
```
